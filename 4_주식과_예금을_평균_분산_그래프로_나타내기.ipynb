```python
# 1. 나눔고딕 폰트 설치
!apt -qq -y install fonts-nanum > /dev/null
!fc-cache -fv > /dev/null

# 2. 라이브러리 설치
!pip install -q finance-datareader

# 3. 라이브러리 불러오기 및 폰트 설정
import FinanceDataReader as fdr
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import pandas as pd
import numpy as np

# 한글 폰트 설정
font_dirs = ['/usr/share/fonts/truetype/nanum']
font_files = fm.findSystemFonts(fontpaths=font_dirs)
for font_file in font_files:
    fm.fontManager.addfont(font_file)
plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['axes.unicode_minus'] = False

# 4. SPY ETF 데이터 불러오기
df = fdr.DataReader('SPY')
df = df[['Adj Close']].rename(columns={'Adj Close': 'SPY'})

# 5. 1년 수익률 계산 (252 거래일 기준)
df['1Y_Return'] = df['SPY'].pct_change(periods=252)

# SPY 수익률 통계
mean_spy = df['1Y_Return'].mean()
std_spy = df['1Y_Return'].std()

# 6. 예금 (연 2%) 수익률 가정
mean_bank = 0.02  # 2%
std_bank = 0.0    # 변동성 없음

# 7. 포트폴리오 (SPY: 70%, 예금: 30%) 수익률 계산
weight_spy = 0.7
weight_bank = 0.3

mean_portfolio = weight_spy * mean_spy + weight_bank * mean_bank
# 분산 = (w^2 * var) + ... 이므로 std = sqrt(w^2 * std^2)
std_portfolio = np.sqrt((weight_spy**2 * std_spy**2) + (weight_bank**2 * std_bank**2))

# 8. 결과 출력
print(f"SPY 평균 수익률: {mean_spy:.2%}, 표준편차: {std_spy:.2%}")
print(f"예금 평균 수익률: {mean_bank:.2%}, 표준편차: {std_bank:.2%}")
print(f"포트폴리오 평균 수익률: {mean_portfolio:.2%}, 표준편차: {std_portfolio:.2%}")

# 9. 시각화
plt.figure(figsize=(8, 8))
plt.title('SPY ETF vs 예금 vs 포트폴리오 (수익률 vs 위험)', fontsize=15)

# 산점도
plt.scatter(std_spy, mean_spy, color='blue', s=100, label='SPY ETF')
plt.scatter(std_bank, mean_bank, color='green', s=100, label='예금 (2%)')
plt.scatter(std_portfolio, mean_portfolio, color='purple', s=100, label='포트폴리오 (7:3)')

# 텍스트 주석 추가
plt.annotate(f'SPY\n({std_spy:.2%}, {mean_spy:.2%})', (std_spy, mean_spy), xytext=(std_spy+0.002, mean_spy+0.002), fontsize=12)
plt.annotate(f'예금\n({std_bank:.2%}, {mean_bank:.2%})', (std_bank, mean_bank), xytext=(std_bank+0.002, mean_bank+0.002), fontsize=12)
plt.annotate(f'포트폴리오\n({std_portfolio:.2%}, {mean_portfolio:.2%})', (std_portfolio, mean_portfolio),
             xytext=(std_portfolio+0.002, mean_portfolio-0.002), fontsize=12)

# 축 설정
plt.xlabel('표준편차 (위험)', fontsize=12)
plt.ylabel('평균 수익률', fontsize=12)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

# SPY - 예금 포트폴리오

```python
# 1. 나눔고딕 폰트 설치
!apt -qq -y install fonts-nanum > /dev/null
!fc-cache -fv > /dev/null

# 2. 필요한 패키지 설치
!pip install -q finance-datareader

# 3. 라이브러리 불러오기
import FinanceDataReader as fdr
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib.ticker as mtick

# 4. 폰트 설정
font_dirs = ['/usr/share/fonts/truetype/nanum']
font_files = fm.findSystemFonts(fontpaths=font_dirs)
for font_file in font_files:
    fm.fontManager.addfont(font_file)
plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['axes.unicode_minus'] = False

# 5. SPY 데이터 불러오기 (전체 거래기간)
df = fdr.DataReader('SPY')
df = df[['Adj Close']].rename(columns={'Adj Close': 'SPY'})

# 6. 1년 수익률 계산 (252 거래일 기준)
df['1Y Return'] = df['SPY'].pct_change(periods=252)
df = df.dropna()

# 7. SPY 평균 수익률 및 표준편차 계산
mean_spy = df['1Y Return'].mean()
std_spy = df['1Y Return'].std()

# 8. 예금 수익률 (2%) 및 위험 (0%)
mean_cash = 0.02
std_cash = 0.0

# 9. 포트폴리오 계산
weights = np.arange(0, 1.01, 0.02)
portfolio_means = []
portfolio_stds = []

for w in weights:
    mean_p = w * mean_spy + (1 - w) * mean_cash
    std_p = np.sqrt((w**2 * std_spy**2) + ((1 - w)**2 * std_cash**2))  # 독립 가정
    portfolio_means.append(mean_p)
    portfolio_stds.append(std_p)

# 10. 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(portfolio_stds, portfolio_means, marker='o', linestyle='-', color='navy')

for i, (x, y) in enumerate(zip(portfolio_stds, portfolio_means)):
    if i == 0:
        label = "예금"
    elif i == len(weights) - 1:
        label = "SPY"
    else:
        label = f"{int(weights[i]*100)}%"
    plt.annotate(label, (x, y), textcoords="offset points", xytext=(0, 5), ha='center', fontsize=9)

plt.gca().xaxis.set_major_formatter(mtick.PercentFormatter(1.0))
plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter(1.0))

plt.xlabel('표준편차 (위험)', fontsize=12)
plt.ylabel('평균 수익률', fontsize=12)
plt.title('SPY-예금 포트폴리오: 수익률 vs 위험', fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.show()
```

3. 전체 거래 기간(상장일 ~ 현재) 데이터를 사용할 것
4. SPY ETF 1년 수익률의 평균과 표준편차를 계산할 것

# 변동성 있는 두 자산에 대해서 분산 투자 하기

```python
# 1. 나눔고딕 폰트 설치 및 설정
!apt-get -qq install -y fonts-nanum > /dev/null
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib as mpl

font_dirs = ['/usr/share/fonts/truetype/nanum']
font_files = fm.findSystemFonts(fontpaths=font_dirs)
for f in font_files:
    fm.fontManager.addfont(f)

plt.rc('font', family='NanumGothic')
mpl.rcParams['axes.unicode_minus'] = False

# 2. FinanceDataReader 설치 및 수정 종가 불러오기
!pip install -q finance-datareader
import FinanceDataReader as fdr
import pandas as pd
import numpy as np

# 전체 기간 SPY, TLT 수정 종가(Adj Close) 불러오기
spy = fdr.DataReader('SPY')
tlt = fdr.DataReader('TLT')

# 수정 종가만 사용
spy_adj = spy['Adj Close']
tlt_adj = tlt['Adj Close']

# 일간 수익률 계산
spy_daily_return = spy_adj.pct_change().dropna()
tlt_daily_return = tlt_adj.pct_change().dropna()

# 1년 = 252 거래일 기준
spy_1y_returns = spy_daily_return.rolling(252).apply(lambda x: (x + 1).prod() - 1)
tlt_1y_returns = tlt_daily_return.rolling(252).apply(lambda x: (x + 1).prod() - 1)

# 수익률 평균과 표준편차 계산
spy_mean = spy_1y_returns.mean()
spy_std = spy_1y_returns.std()
tlt_mean = tlt_1y_returns.mean()
tlt_std = tlt_1y_returns.std()

# 6. 포트폴리오 계산: SPY 투자비율 0% ~ 100% (2% 단위)
weights = np.arange(0, 1.02, 0.02)
port_returns = weights * spy_mean + (1 - weights) * tlt_mean
port_stds = np.sqrt((weights * spy_std) ** 2 + ((1 - weights) * tlt_std) ** 2)

# 7. 그래프 시각화
plt.figure(figsize=(10, 6))
plt.plot(port_stds * 100, port_returns * 100, marker='o', linestyle='-')
plt.xlabel('위험 (표준편차, %)')
plt.ylabel('기대 수익률 (평균, %)')
plt.title('SPY/TLT 분산 포트폴리오 효율성 그래프')
plt.grid(True)

# 8. 포인트 레이블 표시
plt.annotate('TLT', (port_stds[0] * 100, port_returns[0] * 100), fontsize=12, fontweight='bold')
plt.annotate('SPY', (port_stds[-1] * 100, port_returns[-1] * 100), fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```

3. 전체 거래 기간(상장일 ~ 현재) 데이터를 사용할 것
4. SPY ETF 1년 수익률의 평균과 표준편차를 계산할 것
5. TLT ETF 1년 수익률의 평균과 표준편차를 계산할 것
6. SPY ETF와 TLT 분산투자할 것임.
SPY ETF에 대한 투자 비율을 0%부터 100%까지 2% 단위 구간으로 하여 포트폴리오를 계산할 것

# 변동성 있는 두 자산과 예금 포함했을때의 포트폴리오 비교하기

```python
# 1. 나눔고딕 폰트 설치 및 matplotlib 설정
!apt-get -qq install -y fonts-nanum > /dev/null
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib as mpl

# 나눔고딕 등록
font_dirs = ['/usr/share/fonts/truetype/nanum']
font_files = fm.findSystemFonts(fontpaths=font_dirs)
for f in font_files:
    fm.fontManager.addfont(f)

plt.rc('font', family='NanumGothic')
mpl.rcParams['axes.unicode_minus'] = False

# 2. FinanceDataReader 설치 및 수정 종가 불러오기
!pip install -q finance-datareader
import FinanceDataReader as fdr
import pandas as pd
import numpy as np

# 전체 기간 SPY, TLT 수정 종가
spy = fdr.DataReader('SPY')
tlt = fdr.DataReader('TLT')

spy_adj = spy['Adj Close']
tlt_adj = tlt['Adj Close']

# 일간 수익률 계산
spy_ret = spy_adj.pct_change().dropna()
tlt_ret = tlt_adj.pct_change().dropna()

# 누락된 날짜 제거 (공통된 날짜만 사용)
combined = pd.concat([spy_ret, tlt_ret], axis=1, join='inner')
combined.columns = ['SPY', 'TLT']

# 1년 수익률 계산 (252 거래일 기준)
def one_year_return(series):
    return series.rolling(252).apply(lambda x: (x + 1).prod() - 1)

spy_1y = one_year_return(combined['SPY']).dropna()
tlt_1y = one_year_return(combined['TLT']).dropna()

# 4~6. 평균 및 표준편차 계산
spy_mean, spy_std = spy_1y.mean(), spy_1y.std()
tlt_mean, tlt_std = tlt_1y.mean(), tlt_1y.std()

# 예금 2%: 수익률 고정 (평균 = 0.02, 표준편차 = 0)
deposit_mean = 0.02
deposit_std = 0.0

# 9. 투자 비율: 0~100%까지 2% 단위
weights = np.arange(0, 1.02, 0.02)

# 7. SPY-TLT 포트폴리오 계산
port1_returns = weights * spy_mean + (1 - weights) * tlt_mean
port1_stds = np.sqrt((weights * spy_std)**2 + ((1 - weights) * tlt_std)**2)

# 8. SPY-예금 포트폴리오 계산
port2_returns = weights * spy_mean + (1 - weights) * deposit_mean
port2_stds = np.sqrt((weights * spy_std)**2 + ((1 - weights) * deposit_std)**2)

# 9. 그래프 시각화
plt.figure(figsize=(10, 6))

# SPY-TLT
plt.plot(port1_stds * 100, port1_returns * 100, label='SPY-TLT 포트폴리오', color='blue', marker='o')
plt.annotate("TLT", (port1_stds[0]*100, port1_returns[0]*100), fontsize=12, fontweight='bold')
plt.annotate("SPY", (port1_stds[-1]*100, port1_returns[-1]*100), fontsize=12, fontweight='bold')

# SPY-예금
plt.plot(port2_stds * 100, port2_returns * 100, label='SPY-예금 포트폴리오', color='green', marker='s')
plt.annotate("예금", (port2_stds[0]*100, port2_returns[0]*100), fontsize=12, fontweight='bold')
plt.annotate("SPY", (port2_stds[-1]*100, port2_returns[-1]*100), fontsize=12, fontweight='bold')

# 축 라벨 및 설정
plt.xlabel('위험 (표준편차, %)')
plt.ylabel('기대 수익률 (평균, %)')
plt.title('SPY 기반 분산 포트폴리오 비교 (SPY-TLT vs SPY-예금)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 1. 나눔고딕 폰트 설치 및 matplotlib 설정
!apt-get -qq install -y fonts-nanum > /dev/null
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib as mpl

font_dirs = ['/usr/share/fonts/truetype/nanum']
font_files = fm.findSystemFonts(fontpaths=font_dirs)
for f in font_files:
    fm.fontManager.addfont(f)

plt.rc('font', family='NanumGothic')
mpl.rcParams['axes.unicode_minus'] = False

# 2. FinanceDataReader 설치 및 데이터 불러오기
!pip install -q finance-datareader
import FinanceDataReader as fdr
import pandas as pd
import numpy as np

# 전체 기간 SPY, TLT 수정 종가
spy = fdr.DataReader('SPY')
tlt = fdr.DataReader('TLT')
spy_adj = spy['Adj Close']
tlt_adj = tlt['Adj Close']

# 일간 수익률
spy_ret = spy_adj.pct_change().dropna()
tlt_ret = tlt_adj.pct_change().dropna()

# 공통 날짜로 정렬
combined = pd.concat([spy_ret, tlt_ret], axis=1, join='inner')
combined.columns = ['SPY', 'TLT']

# 1년 수익률 (252일 기준)
def one_year_return(series):
    return series.rolling(252).apply(lambda x: (x + 1).prod() - 1)

spy_1y = one_year_return(combined['SPY']).dropna()
tlt_1y = one_year_return(combined['TLT']).dropna()

# 예금 2% 고정 수익률
deposit_1y = pd.Series([0.02] * len(spy_1y), index=spy_1y.index)

# 📊 수익률 분포 그래프 (SPY, TLT, 예금)
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.hist(spy_1y * 100, bins=30, color='skyblue', edgecolor='black')
plt.title('SPY 1년 수익률 분포')
plt.xlabel('수익률 (%)')
plt.ylabel('빈도')

plt.subplot(1, 3, 2)
plt.hist(tlt_1y * 100, bins=30, color='lightgreen', edgecolor='black')
plt.title('TLT 1년 수익률 분포')
plt.xlabel('수익률 (%)')

plt.subplot(1, 3, 3)
plt.hist(deposit_1y * 100, bins=3, color='orange', edgecolor='black')
plt.title('예금 1년 수익률 분포')
plt.xlabel('수익률 (%)')

plt.tight_layout()
plt.show()

# 🔧 평균, 표준편차 계산
spy_mean, spy_std = spy_1y.mean(), spy_1y.std()
tlt_mean, tlt_std = tlt_1y.mean(), tlt_1y.std()
deposit_mean, deposit_std = 0.02, 0.0

# ⚙️ 투자 비율
weights = np.arange(0, 1.02, 0.02)

# 🎯 포트폴리오 성능 계산
port1_returns = weights * spy_mean + (1 - weights) * tlt_mean
port1_stds = np.sqrt((weights * spy_std)**2 + ((1 - weights) * tlt_std)**2)

port2_returns = weights * spy_mean + (1 - weights) * deposit_mean
port2_stds = np.sqrt((weights * spy_std)**2)

# 📈 효율성 그래프
plt.figure(figsize=(10, 6))
plt.plot(port1_stds * 100, port1_returns * 100, label='SPY-TLT 포트폴리오', color='blue', marker='o')
plt.annotate("TLT", (port1_stds[0]*100, port1_returns[0]*100), fontsize=12, fontweight='bold')
plt.annotate("SPY", (port1_stds[-1]*100, port1_returns[-1]*100), fontsize=12, fontweight='bold')

plt.plot(port2_stds * 100, port2_returns * 100, label='SPY-예금 포트폴리오', color='green', marker='s')
plt.annotate("예금", (port2_stds[0]*100, port2_returns[0]*100), fontsize=12, fontweight='bold')
plt.annotate("SPY", (port2_stds[-1]*100, port2_returns[-1]*100), fontsize=12, fontweight='bold')

plt.xlabel('위험 (표준편차, %)')
plt.ylabel('기대 수익률 (평균, %)')
plt.title('SPY 기반 분산 포트폴리오 비교')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# 🎯 각 포트폴리오의 실제 수익률 분포 시뮬레이션 (13번)
port1_dist = []
port2_dist = []

for w in weights:
    p1 = w * spy_1y + (1 - w) * tlt_1y
    p2 = w * spy_1y + (1 - w) * deposit_1y
    port1_dist.append(p1)
    port2_dist.append(p2)

# 🔍 히스토그램 그리기
fig, axes = plt.subplots(2, 5, figsize=(20, 8))
fig.suptitle('포트폴리오별 1년 수익률 분포 (SPY-TLT)', fontsize=16)
for i, ax in enumerate(axes.flat[:len(port1_dist[::10])]):
    w = weights[i*10]
    ax.hist(port1_dist[i*10] * 100, bins=30, color='blue', edgecolor='black')
    ax.set_title(f'SPY {int(w*100)}% / TLT {int((1-w)*100)}%')
    ax.set_xlabel('수익률 (%)')
    ax.set_ylabel('빈도')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()

fig, axes = plt.subplots(2, 5, figsize=(20, 8))
fig.suptitle('포트폴리오별 1년 수익률 분포 (SPY-예금)', fontsize=16)
for i, ax in enumerate(axes.flat[:len(port2_dist[::10])]):
    w = weights[i*10]
    ax.hist(port2_dist[i*10] * 100, bins=30, color='green', edgecolor='black')
    ax.set_title(f'SPY {int(w*100)}% / 예금 {int((1-w)*100)}%')
    ax.set_xlabel('수익률 (%)')
    ax.set_ylabel('빈도')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
